Read file shakes.txt. Use regular expressions in Python  and State Diagrams for each:

1. Change all non-alpha to newlines
2. Sort the data in alphabetical order
3. Merge and count each type
4. Strip the words ending with ing if they have vowel (perform steps in sequence - 1,4,2,3 )
5. Find the sentences which end with !, ?, % and .
6. Find the words which end with !, ?, % and .
7. Calculate the length of word with "."
8. Find the Probabilty of the word with . occurring at the end of the sentence
9. Find the Probabilty of the word with . occurring at the beginning of the sentence

```python
import re
from collections import Counter

with open('shakes.txt', 'r') as file:
    data = file.read()

data_alpha_only = re.sub(r'[^a-zA-Z]', '\n', data)

def strip_ing_vowel(word):
    if word.endswith('ing') and re.search(r'[aeiouAEIOU]', word):
        return word[:-3]
    return word

stripped_words = [strip_ing_vowel(word) for word in data_alpha_only.split()]

sorted_words = sorted(stripped_words)

word_counts = Counter(sorted_words)

sentences = re.findall(r'.*?[!?%.]', data)

words_with_punctuation = re.findall(r'\b\w+[!?%.]', data)

words_with_period = re.findall(r'\b\w+\.', data)
word_lengths_with_period = [len(word) for word in words_with_period]

num_sentences = len(sentences)
num_words_with_period = len(words_with_period)
if num_sentences > 0:
    probability = num_words_with_period / num_sentences
else:
    probability = 0

words_starting_with_period = re.findall(r'\.\w+', data)
num_words_starting_with_period = len(words_starting_with_period)
if num_sentences > 0:
    probability_start = num_words_starting_with_period / num_sentences
else:
    probability_start = 0

print("1. Change all non-alpha to newlines:")
print(data_alpha_only[:1000])  

print("\n2. Sort the data in alphabetical order:")
print(sorted_words[:1000])  

print("\n3. Merge and count each type:")
print(word_counts.most_common(20)) 

print("\n4. Strip the words ending with ing if they have a vowel:")
print(stripped_words[:20])

print("\n5. Find the sentences which end with !, ?, % and .:")
print(sentences[:20]) 

print("\n6. Find the words which end with !, ?, % and .:")
print(words_with_punctuation[:20]) 

print("\n7. Calculate the length of word with \".\":")
print(word_lengths_with_period[:20])  

print("\n8. Find the Probability of the word with . occurring at the end of the sentence:")
print(probability)

print("\n9. Find the Probability of the word with . occurring at the beginning of the sentence:")
print(probability_start)
```
