```python
import nltk
from nltk import pos_tag, RegexpParser
from nltk.tokenize import word_tokenize

# Sample text (replace this with your article text)
text = "Climate change is an urgent global issue that affects the environment in profound ways."

# Tokenize the text
tokens = word_tokenize(text)

# Part-of-speech tagging
tagged_tokens = pos_tag(tokens)

# Define grammar for NP, VP, ADJP, and PP
grammar = r"""
  NP: {<DT|JJ|NN.*>+}          # Noun Phrase
  VP: {<VB.*><NP|PP|CLAUSE>+$}  # Verb Phrase
  ADJP: {<JJ.*><PP>?}           # Adjective Phrase
  PP: {<IN><NP>}                # Prepositional Phrase
"""

# Create a chunk parser
chunk_parser = RegexpParser(grammar)

# Parse the tagged tokens
parsed_tree = chunk_parser.parse(tagged_tokens)

# Extract phrases
noun_phrases = []
verb_phrases = []
adjective_phrases = []
prepositional_phrases = []

for subtree in parsed_tree.subtrees():
    if subtree.label() == 'NP':
        noun_phrases.append(" ".join([token[0] for token in subtree.leaves()]))
    elif subtree.label() == 'VP':
        verb_phrases.append(" ".join([token[0] for token in subtree.leaves()]))
    elif subtree.label() == 'ADJP':
        adjective_phrases.append(" ".join([token[0] for token in subtree.leaves()]))
    elif subtree.label() == 'PP':
        prepositional_phrases.append(" ".join([token[0] for token in subtree.leaves()]))

# Print the extracted phrases
print("Noun Phrases:", noun_phrases)
print("Verb Phrases:", verb_phrases)
print("Adjective Phrases:", adjective_phrases)
print("Prepositional Phrases:", prepositional_phrases)
```
