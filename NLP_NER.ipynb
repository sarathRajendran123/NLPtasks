```python
import os
import nltk
from nltk import pos_tag, ne_chunk
from nltk.tokenize import word_tokenize

# Set up nltk data path
#nltk.data.path.append("/path/to/nltk_data")  # Replace "/path/to/nltk_data" with the actual path to your nltk_data directory

# Path to the bbc folder
bbc_folder = "/content/bbc"

# Get the list of folders inside the bbc folder
folders = os.listdir(bbc_folder)

# Choose the first folder
first_folder = folders[0]

# Get the path to the first folder
first_folder_path = os.path.join(bbc_folder, first_folder)

# Get the list of files inside the first folder
files = os.listdir(first_folder_path)

# Choose the first file
first_file = files[0]

# Get the path to the first file
first_file_path = os.path.join(first_folder_path, first_file)

# Read the content of the first file
with open(first_file_path, 'r') as file:
    content = file.read()

# Tokenize the content
tokens = word_tokenize(content)

# Perform POS tagging
pos_tags = pos_tag(tokens)

print("POS tagged content:")
print(pos_tags)

# Perform named entity recognition
ne_tree = ne_chunk(pos_tags)

print("\nNamed Entities:")
print(ne_tree)
```

```python
import os
import nltk
import pandas as pd
from nltk import pos_tag, ne_chunk
from nltk.tokenize import word_tokenize
from collections import Counter

# Function to perform NER on a file
def ner_on_file(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        text = file.read()
        tokens = word_tokenize(text)
        pos_tags = pos_tag(tokens)
        ne_tree = ne_chunk(pos_tags)
        return ne_tree

# Function to iterate over files in directories
def process_directory(directory):
    ner_results = []
    for root, dirs, files in os.walk(directory):
        for file in files:
            file_path = os.path.join(root, file)
            ner_tree = ner_on_file(file_path)
            ner_results.extend([chunk for chunk in ner_tree if hasattr(chunk, 'label')])
    return ner_results

# Directory containing folders with files
bbc_directory = '/content/bbc'

# Process all files in the directories
all_ner_results = process_directory(bbc_directory)

# Count the occurrences of each named entity
ner_counts = Counter([chunk.label() for chunk in all_ner_results])

# Create a DataFrame to visualize the results
ner_df = pd.DataFrame(ner_counts.items(), columns=['Entity', 'Occurrences'])

print("NER Summary:")
print(ner_df)
```
