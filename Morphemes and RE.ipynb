```python
import re
with open('shakes.txt', 'r') as file:
    text = file.read()

suffix_patterns = {
    'ation': r'\b\w+ation\b',
    'er': r'\b\w+er\b',
    'ness': r'\b\w+ness\b',
    'al': r'\b\w+al\b',
    'able': r'\b\w+able\b',
    'less': r'\b\w+less\b',
    'ity': r'\b\w+ity\b',
    'ize': r'\b\w+ize\b'
}

prefix_patterns = {
    'un': r'\bun\w+\b',
    're': r'\bre\w+\b',
    'in': r'\bin\w+\b',
    'pre': r'\bpre\w+\b',
    'ir': r'\bir\w+\b'
}

def find_words_with_suffixes(text, patterns):
    matching_words = {}
    for suffix, pattern in patterns.items():
        matching_words[suffix] = re.findall(pattern, text)
    return matching_words

def find_words_with_prefixes(text, patterns):
    matching_words = {}
    for prefix, pattern in patterns.items():
        matching_words[prefix] = re.findall(pattern, text)
    return matching_words

words_with_suffixes = find_words_with_suffixes(text, suffix_patterns)
words_with_prefixes = find_words_with_prefixes(text, prefix_patterns)

print("Words ending with specific suffixes:")
for suffix, words in words_with_suffixes.items():
    print(f"{suffix}: {words}")

print("\nWords starting with specific prefixes:")
for prefix, words in words_with_prefixes.items():
    print(f"{prefix}: {words}")
```
