A. Find words with inflectional morpheme as :
1. s
2. es
3. ing
4. ed

```python
import re

with open('shakes.txt', 'r') as file:
    text = file.read()

s_suffix = r'\b\w+s\b'
es_suffix = r'\b\w+es\b'
ing_suffix = r'\b\w+ing\b'
ed_suffix = r'\b\w+ed\b'

words_s_suffix = re.findall(s_suffix, text)
words_es_suffix = re.findall(es_suffix, text)
words_ing_suffix = re.findall(ing_suffix, text)
words_ed_suffix = re.findall(ed_suffix, text)

print("Words with 's' suffix:", words_s_suffix[:30])
print("Words with 'es' suffix:", words_es_suffix[:30])
print("Words with 'ing' suffix:", words_ing_suffix[:30])
print("Words with 'ed' suffix:", words_ed_suffix[:30])
```

B. Find word pairs which are  morphemes of each other in the  file
[For example woman-women, man-men, cut-cat, put-pat, put-cut]

```python
import re

with open('shakes.txt', 'r') as file:
    text = file.read()

def find_morpheme_pairs(text):
    words = re.findall(r'\b\w+\b', text) 
    morpheme_pairs = []

    for word in words:
        if len(word) < 3:
            continue

        morpheme = word[:-1]
        if morpheme in words:
            morpheme_pairs.append((word, morpheme))

    return morpheme_pairs


morpheme_pairs = find_morpheme_pairs(text)

print("Morpheme pairs:")
for pair in morpheme_pairs:
    print(pair[:5])
```

```python
import re

# Define function to check if two words differ by an inflectional morpheme
def are_morphemes_of_each_other(word1, word2):
    inflectional_morphemes = {
        "s": re.compile(r"s$"),
        "es": re.compile(r"es$"),
        "ing": re.compile(r"ing$"),
        "ed": re.compile(r"ed$"),
    }
    # Remove the last character from each word
    base1 = word1[:-1]
    base2 = word2[:-1]
    return base1 == base2 and any(regex.search(word2) for regex in inflectional_morphemes.values())

# Open the file in read mode
with open("shakes.txt", "r") as file:
    # Read the entire file content into a string
    file_content = file.read()

    # Now lowercase and split the content into words
    words = file_content.lower().split()

    # Create a set to store unique pairs
    morpheme_pairs = set()
    for i in range(len(words) - 1):
        for j in range(i + 1, len(words)):
            if are_morphemes_of_each_other(words[i], words[j]):
                pair = sorted([words[i], words[j]])  # Ensure consistent order
                morpheme_pairs.add(tuple(pair))

# Print the unique morpheme pairs
print("Morpheme pairs:")
for pair in morpheme_pairs:
    print(f"- {pair[0]}-{pair[1]}")
```
